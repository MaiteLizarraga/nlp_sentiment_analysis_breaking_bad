```python
# Cargamos el pipeline preentrenado de diarización de hablantes
pipeline = Pipeline.frompretrained("pyannote/speaker-diarization")
```

```python
# Archivo de audio del capítulo
audio_file = "/content/drive/Shareddrives/Capstone/preprocessing_wip_Maite/wav/comprimido/BB1x01.wav"
```

```python
# Realizamos la diarización del archivo de audio
diarization = pipeline(audio_file)
```

```python
# Set para almacenar los diferentes hablantes
unique_speakers = set()
```

```python
# Iteramos sobre los resultados para detectar cuántos hablantes hay
for turn, _, speaker in diarization.itertracks(yield_label=True):
    unique_speakers.add(speaker)  # Añadimos cada hablante identificado al set
```

```python
# Imprimimos el número total de hablantes
num_speakers = len(unique_speakers)
print(f"Number of unique speakers: {num_speakers}")
```
